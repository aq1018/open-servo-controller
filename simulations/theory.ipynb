# Theory

## Motor Model

A typical permanent magnet brushed DC motor can be expressed with the following equations. For a more detailed explanation, or to understand first principles of a brushed DC motor, please familiar yourself with the [DC Motors: Dynamic Model and Control Techniques](https://homepages.laas.fr/lzaccari/seminars/DCmotors.pdf) paper by [Luca Zaccarian](https://homepages.laas.fr/lzaccari/) if you haven't done so already. I will be omitting quite a lot of the basics here.

### Electric Equation

The following equation models the electric properties of a DC motor. This is basically a LR circuit.

$$
  v_{s} = v_{a} + i_{a}R_{a} + L_{a} \frac{di_{a}}{dt}
$$

Where:

$v_{s}$ is the ***supplied voltage***. This is the voltage measured between a voltage power source, such as a battery and ground.

$v_{a}$ is the ***armature voltage*** also known as the ***back EMF voltage***. This is the voltage generated by the motor coil opposing the supplied voltage. This is because when a motor is rotating, it acts as a generator. This value is proportional to the motor speed.

$i_{a}$ is the ***armature current***. This is the current going through the motor coils. This value is proportional to motor torque.

$R_{a}$ is the ***armature resistance***. This is the resistance of the motor coil.

$L_{a}$ is the ***armature inductance***. This is the inductance of the motor coil windings. This prevents current from rising and falling immediately and instead makes motor current follow a exponential decay pattern. Eseentially, a motor coil is a LR circuit.

### Mechanical Equation

This equation approximates the mechanical properties of a DC motor. This incidentally is a good approximation for most of rotating bodies with a damping factor.

$$
T_{a} = T_{L} + J \frac{d\omega}{dt}\ + F\omega
$$

Where:

$T_{a}$ is the ***motor torque***.

$T_{L}$ is the ***load torque***.

$J$ is the ***motor inertia***.

$F$ is the ***viscous friction coefficient***.

### Input Variables

In real world application, we want to provide a PWM signal to drive the motor and control the motor speed. Thus the only input variable in this model is $v_{s}$, the supplied voltage as a function of time.

### Output Variables

We want to find out the following variables in order to provide feedback to the speed (inner) and current/torque ( outer) control loops.

* $\alpha$ - motor angular acceleration.
* $\omega$ - motor angular velocity.
* $\theta$ - motor position.

We will derive the formula to calculate these variables.
